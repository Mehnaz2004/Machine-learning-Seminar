Visualizing the data set to decide which ML model should be used ?

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your data
data = pd.read_csv('student_scores.csv')
feature = data[['Hours']] # Extracts a data-frame object as ML model needs input as a 2D array
target = data['Scores']

# Create the plot
plt.figure(figsize=(8, 5))
plt.scatter(feature, target, color='blue', label='Data points')  # Data points
#plt.plot(feature, lor, color='red', label='Regression line')  # Regression line
plt.title('Hours vs Scores (Linear Regression)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```

As we can see that the points of the scatter-plot follows a Linearity hence we shall opt for a Linear Regressor Model to predict the value of "unseen" data. 

```python
from sklearn.linear_model import LinearRegression

# Train the model
model = LinearRegression()
model.fit(feature.values,target)
```

Model is ready !! Now lets predict some values using ML

```python
# Prediction function
def predict_score(hours):
    hours = float(hours)
    prediction = model.predict([[hours]])
    return prediction[0]

# Example usage
user_input = input("Enter hours studied to predict the score: ")
print(round(predict_score(user_input),2))
```

Why so much of difference in prediction as compared to our data set ?

Spliting the Data Set into Training data & Testing Data to evaluate the models accuracy

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(feature,target)

print(x_train,
      "\n",100*"*","\n",
      x_test,
      "\n",100*"*","\n",
      y_train,
      "\n",100*"*","\n",
      y_test,
      "\n",100*"*","\n")
```

Refitting the model with this trainning data

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

Evaluation of Model Accuracy

```python
model_acc = model.score(x_test,y_test)
print(round(model_acc,2)*100)
```

Lets try to make predictions now ....


```python
# Prediction function
def predict_score(hours):
    hours = float(hours)
    input_data = pd.DataFrame({'Hours': [hours]}) # creating a DataFrame
    prediction = model.predict(input_data)
    return prediction[0]

# Example usage
user_input = input("Enter hours studied to predict the score: ")
print(round(predict_score(user_input),2))
```

Line of Regression (LOR)

```python
lor = model.predict(feature.values)  # point of Regression line
plt.plot(feature, lor, color='red', label='Regression line')  # Regression line
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(feature, target, color='blue', label='Data points')  # Data points
plt.plot(feature, lor, color='red', label='Regression line')  # Regression line
plt.title('Hours vs Scores (Linear Regression)')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.show()
```
