Import the seaborn library 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Read,Extract and Load data from the dataset 

```python
import pandas as pd 

data = pd.read_csv("manufacturing.csv")
print(data)
```

Using Seaborn to visualize the data

```python
# Create a pairplot to visualize relationships between columns
sns.pairplot(data)

# Show the plot
plt.suptitle("Pairplot of Manufacturing Dataset Features", y=1.02, fontsize=16)
plt.show()
```
