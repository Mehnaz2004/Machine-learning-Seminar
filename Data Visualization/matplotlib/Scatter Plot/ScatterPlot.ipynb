Import the matplotlib library 

```python
import matplotlib.pyplot as plt
```

Read,Extract and Load data from the dataset 

```python
import pandas as pd 

data = pd.read_csv("scatter plot.csv")
print(data)

x = data["SAT"]
y = data["GPA"]

print(x)
print(y)
```

Using Matplotlib to visualize the data

```python
plt.figure(figsize=(8, 6)) # Defines the size of the plot
plt.scatter(
    x, y,                  # Plots the points with `x` (SAT scores) on the x-axis and `y` (GPA scores) on the y-axis.
    color='blue',          # Fills the points with a blue color.
    edgecolor='k',         # Adds a black ('k') edge around each point. 'k' is shorthand for black in Matplotlib.
    alpha=0.7             # Sets the transparency of the points to 70% (1 is fully opaque, 0 is fully transparent).   
)

# Add title and labels
plt.title('Scatter Plot of SAT Scores vs GPA', fontsize=14)
plt.xlabel('SAT Scores', fontsize=12)
plt.ylabel('GPA', fontsize=12)

# Customize the grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
